```python
# Map ordered categories
ordinal_mappings = {
    'unresolved_argument': {'none': 0, 'mild': 1, 'serious': 2},
    'work_pressure': {'no': 0, 'minor': 1, 'major': 2},
    'menstrual_phase': {'unknown': 0, 'early': 1, 'mid': 2, 'PMS phase': 3}
}

for col, mapping in ordinal_mappings.items():
    df[col] = df[col].map(mapping)

# Map binary yes/no to 0/1
binary_columns = ['gave_compliments', 'expressed_love', 'self_care', 'ate_fav_food', 'took_bath']
for col in binary_columns:
    df[col] = df[col].map({'no': 0, 'yes': 1})
```

```python
df.head()
```

```python
df['out_with_friends_hrs'].unique()
```

```python
df['out_with_friends_hrs']=df['out_with_friends_hrs'].fillna(df['out_with_friends_hrs'].median())
df['sleep_hours']=df['sleep_hours'].fillna(df['sleep_hours'].median())


# For binary and ordinal categorical columns
for col in ['gave_compliments', 'expressed_love', 'self_care', 'ate_fav_food', 'took_bath',
            'unresolved_argument', 'work_pressure', 'menstrual_phase']:
    df[col]=df[col].fillna(df[col].mode()[0])
```

```python
df.isna().sum()
```

```python
x=df.drop(['mood_score'],axis=1)
x.shape
```

```python
y=df['mood_score']
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual value':y_test, 'Predicted value':y_pred,'Difference':y_test-y_pred})
```

```python
df1
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import r2_score

print("MAE is ",mean_absolute_error(y_test,y_pred))
print("MAPE is ",mean_absolute_percentage_error(y_test,y_pred))
print("MSE is ",mean_squared_error(y_test,y_pred))
print("RMSE is ",root_mean_squared_error(y_test,y_pred))
print("R2 score is ",r2_score(y_test,y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.regplot(x=y_test, y=y_pred, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.xlabel("Actual Mood Score")
plt.ylabel("Predicted Mood Score")
plt.title("Regression Plot: Actual vs Predicted Mood Score")
plt.grid(True)
plt.show()
```
